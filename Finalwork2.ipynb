# Data scientist(Python)

```python
import pandas as pd
```

## Загрузите в колаб файлы по оценкам (ratings) и фильмам (movies) и создайте на их основе pandas-датафреймы

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

## Средствами Pandas, используя dataframe ratings, найдите id пользователя, поставившего больше всего оценок

```python
rating = ratings.groupby('user_id').count().reset_index()
rating_max = rating.sort_values('movie_id', ascending = False).head(1)
rating_max
```

## Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь 

```python
need_user = ratings[ratings['user_id'] == 405]
need_table = need_user.merge(movies, how ='inner', on ='movie_id')
need_table.head()
```

## Добавьте к датафрейму из предыдущего  задания столбцы:

столбцы с общим количеством оценок от всех пользователей на фильм

```python
all_r = ratings.merge(movies, how ='inner', on ='movie_id')
all_r.head()
```

```python
all_count = all_r.groupby('movie_title').count().reset_index()
counts = all_count[['movie_title','user_id']]
counts.columns = ['movie_title','his_count']
counts.head()
```

суммарной оценкой от всех пользователей

```python
all_r.head()
```

```python
all_sum = all_r.groupby('movie_title').sum().reset_index()
sums = all_sum[['movie_title','rating']]
sums.columns = ['movie_title','his_sum']
sums.head()
```

По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет

```python
need_table2 = counts.merge(sums, how ='inner', on ='movie_title')
need_table3 = need_table2.merge(need_table, how ='inner', on ='movie_title')
need_table3.head()
```

```python
table = need_table3[['movie_title', 'his_count', 'his_sum', 'user_id', 'movie_id', 'rating',\
       'release_date', 'Action',\
       'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary',\
       'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',\
       'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']]
```

```python
table.head()
```

## Возьмите модель линейной регрессии (или любую другую для задачи регрессии)  и обучите ее на фильмах

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LinearRegression 
```

```python
X, y = table[['rating', 'Action', 'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary',\
       'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',\
       'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']], table[['his_count', 'his_sum']]
```

## Сформируйте X_train, X_test, y_train, y_test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
# Возвращаем тип DataFrame
X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)
```

```python
# Выбираем тип регрессии для модели
modelL = LinearRegression()
# Обучаем нашу модель
modelL.fit(X_train, y_train)
```

## Оцените качество модели на X_test, y_test при помощи метрик для задачи регрессии

```python
# Оценим качество
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, modelL.predict(X_test)))

from sklearn.metrics import r2_score
print(r2_score(y_test, modelL.predict(X_test)))

from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test, modelL.predict(X_test)))
```

```python
# дополнительная проверка
modelL.score(X_test, y_test)
```
